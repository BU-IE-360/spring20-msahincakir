## Introduction

In this assignment, I try to find an answer to "Are Mortagaged Sales statistics related to mortgage interest rates,exchange rate (USD/TL),and Consumer Price Index?". To avoid complex analysis, I am planning to find only correlation between those time series. The data which I used is provided by CBRT and from March 2017 to March 2020. Also, the frequency of data is monthly.    


I only used xts package.

```r
#install.packages(xts)
library(xts)
```

There are 2 alternatives given here to create a xts object from a csv file. I preferred the second one.

```r
#rawdata <-"https://raw.githubusercontent.com/BU-IE-360/spring20-msahincakir/master/files/hw1/rawdata.csv"
#data_zoo<-read.zoo(rawdata, index.column=0, sep=";",quote = "",format = "%Y-%m-%d")
```

```r
dat<-read.csv('https://raw.githubusercontent.com/BU-IE-360/spring20-msahincakir/master/files/hw1/rawdata.csv',sep=";")
evbs<-xts(dat, order.by = as.Date(rownames(dat), '%Y-%m-%d'))
evbs
```

I calculated the correlation between Mortgaged sales and the other time series which are mortgage interest rates,exchange rate (USD/TL),and Consumer Price Index.

```r
corr<-cor(evbs[,1],evbs[,2:4])
corr
```

As expected, mortgage interest rates is negatively correlated to mortgaged sales.

Then, I plotted each time series in different plots due to different scales of their data. However, the x axis is same, so it is easy follow the relations.

```r
par(mfrow=c(1, 1))
lapply(evbs,function(x){plot(x,main=colnames(x))})
```

## Conclusion

According to results, Mortgaged sales are strongly related with mortgage interest rates as expected. However, CPI and mortagage sales seem not really related wtih each other, which I have not expected. Also, the exchange rate does not have a considerable impact on mortagaged sales.
